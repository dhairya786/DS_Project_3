```python
import pandas as pd

countries = pd.read_csv('https://raw.githubusercontent.com/lukes/ISO-3166-Countries-with-Regional-Codes/master/all/all.csv')
countries_key_fields = countries[['region', 'sub-region', 'name', 'alpha-2']].rename(columns={'name': 'country'})
countries_key_fields.head()

# airports.csv is from https://ourairports.com/data/airports.csv
raw_airports = pd.read_csv('http://ourairports.com/data/airports.csv').rename(columns={'name': 'airport_name'})
main_airports_df = raw_airports[raw_airports.type=="large_airport"][['municipality', 'airport_name', 'latitude_deg', 'longitude_deg', 'elevation_ft', 'iso_country']]
main_airports_df = main_airports_df.merge(countries_key_fields, how='left', left_on='iso_country', right_on='alpha-2')
main_airports_df = main_airports_df.drop_duplicates(['municipality']).dropna()
main_airports_df
```

```python
main_airports_df.to_csv('city_locations.csv', index=False)
```
